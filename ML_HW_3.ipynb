```python
from sklearn.datasets import fetch_california_housing

ca_house_db = fetch_california_housing()

print(ca_house_db.data.shape)
print(ca_house_db.target.shape)
print(ca_house_db.feature_names)
print(ca_house_db.DESCR)
print(ca_house_db)
```

```python
import numpy as np

print(np.isnan(ca_house_db.data).any())
print(np.isinf(ca_house_db.data).any())
print(np.isnan(ca_house_db.target).any())
print(np.isinf(ca_house_db.target).any())
```

# Question 1

```python
from sklearn.model_selection import train_test_split

X_train1, X_test, y_train1, y_test = train_test_split(ca_house_db.data, ca_house_db.target, test_size=0.5,
                                                      random_state=42)
X_train2, X_val, y_train2, y_val = train_test_split(X_train1, y_train1, test_size=0.2, random_state=42)

print(X_train1.shape, y_train1.shape)
print(X_train2.shape, y_train2.shape)
print(X_val.shape, y_val.shape)
print(X_test.shape, y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

def create_model(num_hidden_layers, num_units, optimizer, activation, output_function):
    # Create a new sequential model
    model = Sequential()
    # Add the input layer with the specified number of units and activation function
    model.add(Dense(num_units, input_shape=(X_train2.shape[1],), activation=activation))
    # Add the specified number of hidden layers with the specified number of units and activation function
    for i in range(num_hidden_layers):
        model.add(Dense(num_units, activation=activation))
    # Add the output layer with a single unit and the specified output activation function
    model.add(Dense(1, activation=output_function))
    # Compile the model with the mean squared error loss, the specified optimizer, and mean squared error metric
    model.compile(loss='mean_squared_error', optimizer=optimizer, metrics=['mse'])
    # Return the compiled model
    return model
```

```python
import numpy as np

# set of hyperparameters 1
num_hidden_layers_1 = 2
num_units_1 = 32
optimizer_1 = 'adam'
activation_1 = 'tanh'
output_function_1 = 'linear'

model_1 = create_model(num_hidden_layers_1, num_units_1, optimizer_1, activation_1, output_function_1)
history_1 = model_1.fit(X_train2, y_train2, epochs=100, batch_size=32, validation_data=(X_val, y_val))

# set of hyperparameters 2
num_hidden_layers_2 = 3
num_units_2 = 64
optimizer_2 = 'adam'
activation_2 = 'sigmoid'
output_function_2 = 'linear'

model_2 = create_model(num_hidden_layers_2, num_units_2, optimizer_2, activation_2, output_function_2)
history_2 = model_2.fit(X_train2, y_train2, epochs=100, batch_size=32, validation_data=(X_val, y_val))

# set of hyperparameters 3
num_hidden_layers_3 = 1
num_units_3 = 16
optimizer_3 = 'adam'
activation_3 = 'relu'
output_function_3 = 'linear'

model_3 = create_model(num_hidden_layers_3, num_units_3, optimizer_3, activation_3, output_function_3)
history_3 = model_3.fit(X_train2, y_train2, epochs=100, batch_size=32, validation_data=(X_val, y_val))
```

```python
X_test
```

```python
from sklearn.metrics import r2_score

# Evaluate model 1
y_pred_1 = model_1.predict(X_val)
r2_1 = r2_score(y_val, y_pred_1)
print("R2 score for model 1: {:.3f}".format(r2_1))

# Evaluate model 2
y_pred_2 = model_2.predict(X_val)
r2_2 = r2_score(y_val, y_pred_2)
print("R2 score for model 2: {:.3f}".format(r2_2))

# Evaluate model 3
y_pred_3 = model_3.predict(X_val)
r2_3 = r2_score(y_val, y_pred_3)
print("R2 score for model 3: {:.3f}".format(r2_3))
```

# Therefore, the model with high R2 is Model 2. It's R2 score is 63.5% that is way much better than Model 1 and Model 3. 

# Question 3

```python
y_pred_test = model_2.predict(X_test)

# Calculate R2 score on testing data
r2_test = r2_score(y_test, y_pred_test)

# Print R2 score rounded to 2 decimal places
print("R2 score on testing data:", round(r2_test, 3))
```

This model is consistent with X_test data as well. That the Model 2 has the same R2 score on all the testing data. 

# Question 4

```python
# Calculate the absolute error for each testing sample
abs_errors = np.abs(y_pred_test - y_test)

# Sort the samples by absolute error
sorted_indices = np.argsort(abs_errors, axis=0)[::-1]

# Print the ten samples with the largest absolute errors
for i in range(10):
    index = sorted_indices[i]
    print(f'Sample {index}: predicted value = {y_pred_test[index]}, actual value = {y_test[index]}, absolute error = {abs_errors[index]}')
```
